```python
import pandas as pd
import numpy as np
import os
import zipfile
from IPython.display import display

# Supaya tampilan tabel lebih rapi
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

print("Library berhasil di-import.")
```

```python
daftar_file_csv = [file for file in os.listdir() if file.endswith(".csv")]

if len(daftar_file_csv) == 0:
    raise FileNotFoundError("Tidak ada file CSV di Colab")

# Mengutamakan file yang mengandung kata 'MTA'
file_mta = [file for file in daftar_file_csv if "MTA" in file or "mta" in file]

if len(file_mta) > 0:
    file_path = file_mta[0]
else:
    file_path = daftar_file_csv[0]

df_raw = pd.read_csv(file_path)
print("Dataset berhasil dibaca.")
print("Nama file:", file_path)
print("Ukuran dataset:", df_raw.shape)

display(df_raw.head())
```

```python
# CEK STRUKTUR AWAL DATASET

print("Jumlah baris:", df_raw.shape[0])
print("Jumlah kolom:", df_raw.shape[1])

print("\nDaftar kolom:")
for kolom in df_raw.columns:
    print("-", kolom)

print("\nTipe data awal:")
display(df_raw.dtypes.to_frame("tipe_data"))

print("\nJumlah missing value awal:")
display(df_raw.isnull().sum().to_frame("jumlah_missing"))

print("\nJumlah duplikasi baris:", df_raw.duplicated().sum())

if "Date" in df_raw.columns:
    print("Jumlah duplikasi tanggal:", df_raw["Date"].duplicated().sum())
else:
    print("Kolom Date tidak ditemukan.")
```

```python
# VALIDASI KOLOM DATASET

kolom_wajib = [
    "Date",
    "Subways: Total Estimated Ridership",
    "Subways: % of Comparable Pre-Pandemic Day",
    "Buses: Total Estimated Ridership",
    "Buses: % of Comparable Pre-Pandemic Day",
    "LIRR: Total Estimated Ridership",
    "LIRR: % of Comparable Pre-Pandemic Day",
    "Metro-North: Total Estimated Ridership",
    "Metro-North: % of Comparable Pre-Pandemic Day",
    "Access-A-Ride: Total Scheduled Trips",
    "Access-A-Ride: % of Comparable Pre-Pandemic Day",
    "Bridges and Tunnels: Total Traffic",
    "Bridges and Tunnels: % of Comparable Pre-Pandemic Day",
    "Staten Island Railway: Total Estimated Ridership",
    "Staten Island Railway: % of Comparable Pre-Pandemic Day"
]

kolom_tidak_ada = [kolom for kolom in kolom_wajib if kolom not in df_raw.columns]

if len(kolom_tidak_ada) > 0:
    print("Ada kolom yang tidak ditemukan:")
    for kolom in kolom_tidak_ada:
        print("-", kolom)
    raise ValueError("Nama kolom dataset tidak sesuai. Cek kembali file CSV.")
else:
    print("Semua kolom wajib tersedia. Dataset siap diproses.")
```

```python
# CLEANING DATA
df_clean = df_raw.copy()

# Date menjadi datetime
df_clean["Date"] = pd.to_datetime(df_clean["Date"], format="%m/%d/%Y", errors="coerce")

jumlah_tanggal_error = df_clean["Date"].isnull().sum()

if jumlah_tanggal_error > 0:
    raise ValueError(f"Ada {jumlah_tanggal_error} tanggal yang gagal dibaca")

# Fungsi untuk membersihkan angka
def bersihkan_angka(nilai):
    """
    Fungsi ini mengubah angka berbentuk teks menjadi numerik.
    Contoh: '2,212,965' menjadi 2212965
    """
    if pd.isna(nilai):
        return np.nan

    nilai = str(nilai)
    nilai = nilai.replace(",", "")
    nilai = nilai.strip()

    return pd.to_numeric(nilai, errors="coerce")


# Fungsi untuk membersihkan persen
def bersihkan_persen(nilai):
    """
    Fungsi ini mengubah persen berbentuk teks menjadi numerik.
    Contoh: '97%' menjadi 97
    """
    if pd.isna(nilai):
        return np.nan

    nilai = str(nilai)
    nilai = nilai.replace("%", "")
    nilai = nilai.strip()

    return pd.to_numeric(nilai, errors="coerce")


# Kolom jumlah ridership / traffic / scheduled trips
kolom_ridership = [
    "Subways: Total Estimated Ridership",
    "Buses: Total Estimated Ridership",
    "LIRR: Total Estimated Ridership",
    "Metro-North: Total Estimated Ridership",
    "Access-A-Ride: Total Scheduled Trips",
    "Bridges and Tunnels: Total Traffic",
    "Staten Island Railway: Total Estimated Ridership"
]

# Kolom persentase terhadap hari pra-pandemi
kolom_persen = [
    "Subways: % of Comparable Pre-Pandemic Day",
    "Buses: % of Comparable Pre-Pandemic Day",
    "LIRR: % of Comparable Pre-Pandemic Day",
    "Metro-North: % of Comparable Pre-Pandemic Day",
    "Access-A-Ride: % of Comparable Pre-Pandemic Day",
    "Bridges and Tunnels: % of Comparable Pre-Pandemic Day",
    "Staten Island Railway: % of Comparable Pre-Pandemic Day"
]

for kolom in kolom_ridership:
    df_clean[kolom] = df_clean[kolom].apply(bersihkan_angka)

for kolom in kolom_persen:
    df_clean[kolom] = df_clean[kolom].apply(bersihkan_persen)

# Menambahkan atribut waktu untuk kebutuhan Data Warehouse
df_clean["date_key"] = df_clean["Date"].dt.strftime("%Y%m%d").astype(int)
df_clean["year"] = df_clean["Date"].dt.year
df_clean["quarter"] = df_clean["Date"].dt.quarter
df_clean["month"] = df_clean["Date"].dt.month
df_clean["month_name"] = df_clean["Date"].dt.month_name()
df_clean["day"] = df_clean["Date"].dt.day
df_clean["day_name"] = df_clean["Date"].dt.day_name()

print("Cleaning data selesai.")
print("Ukuran data setelah cleaning:", df_clean.shape)

display(df_clean.head())
```

```python
# VALIDASI SETELAH CLEANING

print("Jumlah baris:", df_clean.shape[0])
print("Jumlah kolom:", df_clean.shape[1])

print("\nRentang tanggal dataset:")
print("Tanggal awal :", df_clean["Date"].min())
print("Tanggal akhir:", df_clean["Date"].max())

print("\nJumlah missing value setelah cleaning:")
display(df_clean.isnull().sum().to_frame("jumlah_missing"))

print("\nTipe data setelah cleaning:")
display(df_clean.dtypes.to_frame("tipe_data"))
```

```python
# TRANSFORMASI WIDE KE LONG

# Mapping kolom setiap moda transportasi
mapping_moda = {
    "Subways": {
        "ridership_col": "Subways: Total Estimated Ridership",
        "recovery_col": "Subways: % of Comparable Pre-Pandemic Day",
        "value_type": "Estimated Ridership"
    },
    "Buses": {
        "ridership_col": "Buses: Total Estimated Ridership",
        "recovery_col": "Buses: % of Comparable Pre-Pandemic Day",
        "value_type": "Estimated Ridership"
    },
    "LIRR": {
        "ridership_col": "LIRR: Total Estimated Ridership",
        "recovery_col": "LIRR: % of Comparable Pre-Pandemic Day",
        "value_type": "Estimated Ridership"
    },
    "Metro-North": {
        "ridership_col": "Metro-North: Total Estimated Ridership",
        "recovery_col": "Metro-North: % of Comparable Pre-Pandemic Day",
        "value_type": "Estimated Ridership"
    },
    "Access-A-Ride": {
        "ridership_col": "Access-A-Ride: Total Scheduled Trips",
        "recovery_col": "Access-A-Ride: % of Comparable Pre-Pandemic Day",
        "value_type": "Scheduled Trips"
    },
    "Bridges and Tunnels": {
        "ridership_col": "Bridges and Tunnels: Total Traffic",
        "recovery_col": "Bridges and Tunnels: % of Comparable Pre-Pandemic Day",
        "value_type": "Total Traffic"
    },
    "Staten Island Railway": {
        "ridership_col": "Staten Island Railway: Total Estimated Ridership",
        "recovery_col": "Staten Island Railway: % of Comparable Pre-Pandemic Day",
        "value_type": "Estimated Ridership"
    }
}

list_data_long = []

for moda, info in mapping_moda.items():
    temp = df_clean[[
        "Date",
        "date_key",
        "year",
        "quarter",
        "month",
        "month_name",
        "day",
        "day_name"
    ]].copy()

    temp["transport_mode"] = moda
    temp["value_type"] = info["value_type"]
    temp["ridership_value"] = df_clean[info["ridership_col"]]
    temp["recovery_pct"] = df_clean[info["recovery_col"]]

    list_data_long.append(temp)

# Menggabungkan seluruh moda menjadi satu tabel long
df_long = pd.concat(list_data_long, ignore_index=True)

df_long = df_long.rename(columns={"Date": "date"})

# Mengurutkan data berdasarkan tanggal dan moda
df_long = df_long.sort_values(["date", "transport_mode"]).reset_index(drop=True)

print("Transformasi wide ke long selesai.")
print("Ukuran data long:", df_long.shape)

display(df_long.head(10))
```

```python
# VALIDASI DATA LONG

print("Jumlah baris data long:", len(df_long))
print("Jumlah kolom data long:", df_long.shape[1])
print("Jumlah moda transportasi:", df_long["transport_mode"].nunique())

print("\nJumlah data per moda:")
display(df_long["transport_mode"].value_counts().to_frame("jumlah_record"))

print("\nRentang tanggal data long:")
print("Tanggal awal :", df_long["date"].min())
print("Tanggal akhir:", df_long["date"].max())

print("\nJumlah missing value data long:")
display(df_long.isnull().sum().to_frame("jumlah_missing"))

# Validasi jumlah baris
jumlah_seharusnya = len(df_clean) * df_long["transport_mode"].nunique()
jumlah_aktual = len(df_long)

print("\nCek jumlah baris:")
print("Jumlah seharusnya:", jumlah_seharusnya)
print("Jumlah aktual    :", jumlah_aktual)

if jumlah_seharusnya == jumlah_aktual:
    print("Validasi berhasil. Jumlah baris data long sudah sesuai.")
else:
    print("Validasi gagal. Jumlah baris data long belum sesuai.")
```

```python
# MEMBUAT DIMENSION & FACT TABLE

# Membuat dimensi tanggal
dim_date = df_long[[
    "date_key",
    "date",
    "year",
    "quarter",
    "month",
    "month_name",
    "day",
    "day_name"
]].drop_duplicates().copy()

dim_date = dim_date.sort_values("date_key").reset_index(drop=True)

# Membuat dimensi moda transportasi
dim_transport_mode = df_long[[
    "transport_mode",
    "value_type"
]].drop_duplicates().copy()

dim_transport_mode = dim_transport_mode.sort_values("transport_mode").reset_index(drop=True)
dim_transport_mode["mode_id"] = range(1, len(dim_transport_mode) + 1)

dim_transport_mode = dim_transport_mode[[
    "mode_id",
    "transport_mode",
    "value_type"
]]

# Membuat fact table
fact_daily_ridership = df_long.merge(
    dim_transport_mode[["mode_id", "transport_mode"]],
    on="transport_mode",
    how="left"
)

fact_daily_ridership = fact_daily_ridership[[
    "date_key",
    "mode_id",
    "ridership_value",
    "recovery_pct"
]].copy()

# Menambahkan primary key sederhana untuk fact table
fact_daily_ridership.insert(0, "fact_id", range(1, len(fact_daily_ridership) + 1))

print("Dimensi tanggal:")
display(dim_date.head())

print("\nDimensi moda transportasi:")
display(dim_transport_mode)

print("\nFact daily ridership:")
display(fact_daily_ridership.head())

print("\nUkuran dim_date:", dim_date.shape)
print("Ukuran dim_transport_mode:", dim_transport_mode.shape)
print("Ukuran fact_daily_ridership:", fact_daily_ridership.shape)
```

```python
# EXPORT FILE

folder_output = "output_transformasi_mta"
os.makedirs(folder_output, exist_ok=True)

# Ringkasan Dataset
ringkasan_dataset = pd.DataFrame({
    "aspek": [
        "Jumlah baris data awal",
        "Jumlah kolom data awal",
        "Tanggal awal",
        "Tanggal akhir",
        "Jumlah moda transportasi",
        "Jumlah baris data long",
        "Jumlah baris dim_date",
        "Jumlah baris dim_transport_mode",
        "Jumlah baris fact_daily_ridership",
        "Jumlah missing value data long"
    ],
    "nilai": [
        df_raw.shape[0],
        df_raw.shape[1],
        df_long["date"].min(),
        df_long["date"].max(),
        df_long["transport_mode"].nunique(),
        df_long.shape[0],
        dim_date.shape[0],
        dim_transport_mode.shape[0],
        fact_daily_ridership.shape[0],
        df_long.isnull().sum().sum()
    ]
})

# Export data hasil cleaning dan transformasi
df_clean.to_csv(f"{folder_output}/mta_clean_wide.csv", index=False)
df_long.to_csv(f"{folder_output}/mta_long_clean.csv", index=False)

# Export tabel dimensi dan fakta
dim_date.to_csv(f"{folder_output}/dim_date.csv", index=False)
dim_transport_mode.to_csv(f"{folder_output}/dim_transport_mode.csv", index=False)
fact_daily_ridership.to_csv(f"{folder_output}/fact_daily_ridership.csv", index=False)

# Export ringkasan dataset
ringkasan_dataset.to_csv(f"{folder_output}/ringkasan_dataset.csv", index=False)

print("File berhasil diexport ke folder:", folder_output)

print("\nDaftar file output:")
for nama_file in os.listdir(folder_output):
    print("-", nama_file)
```
